```python
import pylab
import boolean2
from boolean2 import state, network, util, Model 
```

```python
text = """
#initial values 
pheromone = pdp-1 = False
cmk-1 = ncr = nhr-8 = True
swi-snf = ifta-2 = ftt-1 = False
rle-1 = par-5 = hlh-30 = True
daf-18 = ins-1 = bra-1 = False

#initial conditions
aap-1 *= daf-2 
age-1 *= aap-1 and not daf-18 
akt *= pdk-1 
daf-11 *= daf-25 or (not srbc and not gpa-3) 
daf-2 *= (ins-7 or daf-28 or ins-6) and (not ins-1 and not ins-18)
daf-12 *= not daf-9 
daf-21 *= daf-21 or daf-2 
daf-3 *= not daf-8-14 and not daf-7 
daf-36 *= nhr-8 
daf-5 *= daf-3 
din-1s *= daf-12 
pdk-1 *= age-1 
tax-4 *= daf-11 
ins-7 *= (daf-7 or ins-35-33) and not daf-16
daf-8-14 *= daf-1-4 and not pdp-1 
daf-9 *= daf-12 or not dhs-16 or not daf-36 or ncr or hsf-1 
daf-7 *= (daf-11 or tax-4 or cmk-1) and (not pheromone and not hsf-1) 
daf-28 *= daf-7 or daf-11 or cmk-1 or not ins-6 
daf-16 *= (daf-12 or daf-5 or not akt or daf-18) and (swi-snf and ftt-1 or ifta-2) and (not par-5 or not rle-1 or not hlh-30) 
ins-18 *= daf-16  
dauer *= daf-16 or din-1s 
srbc *= pheromone 
tph-1 *= (daf-3) or not tax-4 or not daf-7 
gpa-3 *= srbc 
daf-15 *= not daf-16 
dhs-16 *= daf-2 
daf-1-4 *= (daf-7 or daf-21) and not bra-1 
ins-4 *= not daf-28 
ins-6 *= not ins-4 or cmk-1
ins-35-33 *= not ins-6 
hsf-1 *= not daf-2
"""
```

```python
coll = util.Collector()
for i in range(100):
    sim = Model(mode='async', text=text)
    sim.initialize(missing= util.randbool)
    sim.iterate(steps=200)
    coll.collect( states=sim.states, nodes=sim.nodes )

avgs = coll.get_averages( normalize=True )
# print("ins-6",avgs["ins-6"])
print("daf-7",avgs["daf-7"],"-----------------")
print("daf-11",avgs["daf-11"],"-----------------")
print("daf-7",avgs["daf-7"], "-----------------")
print("daf-9",avgs["daf-9"], "-----------------")

print("tax-4",avgs["tax-4"],"-----------------")
print("hsf-1",avgs["hsf-1"],"-----------------")
print("cmk-1",avgs["cmk-1"],"-----------------")

print("daf-2",avgs["daf-2"],"-----------------")
print("daf-11",avgs["daf-11"],"-----------------")
print("ins-4",avgs["ins-4"],"-----------------")
print("daf-28",avgs["daf-28"],"-----------------")
print("pheromone",avgs["pheromone"],"-----------------")
print("dauer",avgs["dauer"], "-----------------")
print("daf-16",avgs["daf-16"], "-----------------")

#for state in sim.states:
 #   print(state.pheromone)
#sim.report_cycles()
#sim.detect_cycles()
```
